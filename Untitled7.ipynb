```python
# 1. Install & Import Libraries
!pip install faker

import pandas as pd
import numpy as np
from faker import Faker
import random
import matplotlib.pyplot as plt

# 2. Create Fake Dataset using Faker
fake = Faker()

data = []

for _ in range(500):   # number of rows
    date = fake.date_between(start_date='-2y', end_date='today')
    country = fake.country()

    confirmed = random.randint(1000, 100000)
    deaths = random.randint(0, int(confirmed * 0.1))
    recovered = random.randint(0, confirmed - deaths)
    active = confirmed - (deaths + recovered)

    data.append([date, country, confirmed, deaths, recovered, active])

df = pd.DataFrame(data, columns=[
    'Date', 'Country', 'Confirmed', 'Deaths', 'Recovered', 'Active'
])

df.head()

# 3. Data Preprocessing
df['Date'] = pd.to_datetime(df['Date'])

# Check missing values
print(df.isnull().sum())

# Remove duplicates
df.drop_duplicates(inplace=True)

# 4. Basic Analysis
print("Total Confirmed:", df['Confirmed'].sum())
print("Total Deaths:", df['Deaths'].sum())
print("Total Recovered:", df['Recovered'].sum())

# 5. Group by Country
country_data = df.groupby('Country')[['Confirmed', 'Deaths', 'Recovered']].sum()
country_data = country_data.sort_values(by='Confirmed', ascending=False)

country_data.head()

# 6. Visualization (Matplotlib)
# Line Chart (Trend Over Time)
daily = df.groupby('Date')['Confirmed'].sum()

plt.figure()
plt.plot(daily)
plt.title("Confirmed Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Cases")
plt.show()

# Bar Chart (Top Countries)
top_countries = country_data.head(10)

plt.figure()
plt.bar(top_countries.index, top_countries['Confirmed'])
plt.xticks(rotation=90)
plt.title("Top 10 Countries - Confirmed Cases")
plt.show()

# Pie Chart
totals = [
    df['Confirmed'].sum(),
    df['Deaths'].sum(),
    df['Recovered'].sum()
]

labels = ['Confirmed', 'Deaths', 'Recovered']

plt.figure()
plt.pie(totals, labels=labels, autopct='%1.1f%%')
plt.title("Case Distribution")
plt.show()

# 7. Correlation Heatmap
import seaborn as sns

corr = df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr()

plt.figure()
sns.heatmap(corr, annot=True)
plt.title("Correlation Matrix")
plt.show()
```
